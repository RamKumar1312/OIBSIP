# Loading Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

# Loading Dataset

```python
columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Class_labels']
#Loading the data
df = pd.read_csv('Iris.csv',names=columns)
df.head()
```

# Visualization

```python
df.describe()
```

```python
sns.pairplot(df, hue='Class_labels')
```

# Input and Output Column Seperation

```python
data = df.values

A = data[:,0:4]
B = data[:,4]
```

# Splitting The Training and Test Data

```python
from sklearn.model_selection import train_test_split
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.2)
```

# Using SVM (Support Vector Machine)

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(A_train, B_train)
```

```python
prediction_1 = model_svc.predict(A_test)
#Calculating accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(B_test, prediction_1)*100)
```

By using **Support Vector Machine Algorithm**, the accuracy score percentage for the classification is 100.0

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(A_train, B_train)
```

```python
prediction_2 = model_LR.predict(A_test)
#Calculating Accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(B_test, prediction_2)*100)
```

Using **Logistic Regression**, the accuracy score for this type of classification is 96.67

# Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model_RF=RandomForestClassifier()
model_RF.fit(A_train,B_train)
```

```python
prediction_3 = model_RF.predict(A_test)
#Calculating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(B_test, prediction_3)*100)
```

Using Random Forest Classifier Algorithm, the accuracy score for the classification of the dataset values is 96.67

# Using Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc=DecisionTreeClassifier()
model_dtc.fit(A_train, B_train)
```

```python
prediction_4 = model_dtc.predict(A_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(B_test, prediction_4)*100)
```

By using Decision Tree Classifier Algorithm Too, we get the same accuracy score of 96.67 which is similiar to so far passed outcomes.
