# Unemployment Analysis Using Python

We use python to visualize and analyze the unemployment data of all the states of India, which could make a clear view of several illiteracy regions of the nation.

##  Data Analysis

```python
data.States.unique()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=data.Date,y=data['Estimated Unemployment Rate'])
plt.title("Unemployment rate in India over time")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
sns.barplot(data,x='Region',y='Estimated Unemployment Rate')
plt.xlabel("Region")
plt.ylabel("Average Unemployment Rate")
plt.title('Unemployment Rate By Region')
plt.show()
```

```python
data.groupby('Region')['Estimated Unemployment Rate'].mean().sort_values(ascending=False)
```

From these numbers we could see that the North Region has the highest illiteracy and unemployment rate, while the west region has the lowest unemployment rate.

```python
plt.figure(figsize=(12,8))
sns.barplot(data,x='States',y='Estimated Unemployment Rate')
plt.ylabel('Average Unemployment Rate')
plt.xticks(rotation=90)
plt.show()
```

* We can see that Haryana and Tripura have the highest unemployment rate, while Meghalaya and Assam have the lowest unemployment rate.



```python
for region in list(data['Region'].unique()):
  df=data[data['Region']==region]
  rate_by_states=df.groupby('States')['Estimated Unemployment Rate'].mean().sort_values(ascending=False)
  print(f'-> {rate_by_states.index[0]} has the highest unemployment rate of {rate_by_states[0]}% in {region} region.\n')
```
